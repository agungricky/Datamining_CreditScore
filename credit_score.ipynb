```python
# Mengisi nilai yang hilang dengan mean
mean_values = df.fillna(df.mean(numeric_only=True))

# Menghitung perbedaan antara data asli dan data yang telah diganti dengan mean pada setiap kolom
perbedaan_mean = (df != mean_values).sum().sum()

# Menghitung akurasi menggunakan mean
akurasi_mean = 1 - (perbedaan_mean / (df.shape[0] * df.shape[1]))

# Menghitung true positives menggunakan mean
true_positives_mean = ((df.notnull()) & (mean_values.notnull())).sum().sum()

# Menghitung false positives menggunakan mean
false_positives_mean = (df.isnull() & (mean_values.notnull())).sum().sum()

# Menghitung presisi menggunakan mean
presisi_mean = true_positives_mean / (true_positives_mean + false_positives_mean)

# Menghitung recall menggunakan mean
recall_mean = 1 - (mean_values.isnull().sum().sum() / (df.shape[0] * df.shape[1]))

# Menghitung specificity menggunakan mean
specificity_mean = 1 - (df.isnull().sum().sum() / (df.shape[0] * df.shape[1]))

# Menghitung f-score menggunakan mean
f_score_mean = 2 * (presisi_mean * recall_mean) / (presisi_mean + recall_mean)

# Cek jumlah missing value setelah menggunakan mean
print("Jumlah missing value setelah pengisian dengan mean:")
hasil_mean = mean_values.copy()  # Membuat salinan data dengan pengolahan median
print(hasil_mean.isnull().sum())
```

```python
#tangani missing value dengan median
median_values = df.fillna(df.median(numeric_only=True))

# Menghitung perbedaan antara data asli dan data yang telah diganti dengan mean pada setiap kolom
perbedaan = (df != median_values).sum().sum()

# Menghitung akurasi
akurasi = 1 - (perbedaan / (df.shape[0] * df.shape[1]))

# Menghitung presisi
presisi = 1 - (perbedaan / median_values.size)

# Menghitung recall
recall = 1 - akurasi

# Menghitung Specificity
specificity = 1 - (perbedaan / (df.shape[0] * df.shape[1]))

# Menghitung F-Score
beta = 1  # Nilai beta untuk F-Score (F1-Score)
f_score = (1 + beta*2) * (presisi * recall) / ((beta*2 * presisi) + recall)

print("Akurasi: {:.2%}".format(akurasi))
print("Presisi: {:.2%}".format(presisi))
print("Recall: {:.2%}".format(recall))
print("Specificity: {:.2%}".format(specificity))
print("F-Score: {:.2%}".format(f_score))

# Mengupdate dataset dengan nilai yang telah diisi
df.update(median_values)

#cek jumlah missing value
median_values.isnull().sum()
```
